```python
#Eliminate if not dog
def f(row):
    if (((row['p1_conf'])>(row['p2_conf']))|((row['p1_conf'])>(row['p3_conf'])) and (row['p1_dog']!=False)):
        val=row['p1']
    elif (((row['p2_conf'])>(row['p1_conf']))|((row['p2_conf'])>(row['p3_conf'])) and (row['p2_dog']!=False)):
        val=row['p2']
    elif (((row['p3_conf'])>(row['p1_conf']))|((row['p3_conf'])>(row['p2_conf'])) and (row['p2_dog']!=False)):
        val=row['p3']
    else:
        val="not dog"
    return val
#https://stackoverflow.com/questions/21702342/creating-a-new-column-based-on-if-elif-else-condition
```

```python
df_b_clean['Dog_type']=df_b_clean.apply(f,axis=1)
```

```python
#drop unnecessary 
df_b_clean=df_b_clean.drop(columns=['p1','p1_conf','p2','p2_conf','p3','p3_conf','p1_dog','p2_dog','p3_dog'],axis=1)
```

```python
df_b_clean.head()
```

```python
df_b_clean=df_b_clean[df_b_clean['rating_numerator']!=0]
df_b_clean=df_b_clean[df_b_clean['Dog_type']!="not dog"]
```
